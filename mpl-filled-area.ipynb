```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import seaborn as sns
%matplotlib notebook
```

```python
df = pd.read_csv('Canadian_climate_history.csv', index_col=0, parse_dates=[0])
df.dropna(inplace=True)
df.head()
```

```python
x = df.index
y1 = 'TOTAL_PRECIPITATION_STJOHNS'
y2 = 'TOTAL_PRECIPITATION_TORONTO'
y3 = 'TOTAL_PRECIPITATION_MONTREAL'
title = 'Total precipitation in '

facecolor = '#f0eeee' 
font_color = '#525252'
hfont = {'fontname':'Calibri'}

sns.set_style('whitegrid')
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, sharex=True, facecolor=facecolor)
fig.set_size_inches(9, 6)

ax1.fill_between(x, y1, data=df, color='#F79B77', linewidth=2)
ax1.set_title(title+'St. John\'s', fontsize=18, pad=10, color=font_color, **hfont)

ax2.fill_between(x, y2, data=df, color='#01B8AA', linewidth=2)
ax2.set_title(title+'Toronto', fontsize=18, pad=10, color=font_color, **hfont)

ax3.fill_between(x, y3, data=df, color='#1B4B5A', linewidth=2)
ax3.set_title(title+'Montreal', fontsize=18, pad=10, color=font_color, **hfont)
ax3.set_xlabel('Year', fontsize=12, color=font_color, **hfont)

for ax in (ax1, ax2, ax3):
    ax.set_ylabel('mm', fontsize=12, color=font_color, **hfont)
    ax.set_yticks([0, 50, 100, 150])

for label in (ax1.get_yticklabels() + ax2.get_yticklabels()):
    label.set(fontsize=12, color=font_color, **hfont)
for label in (ax3.get_xticklabels() + ax3.get_yticklabels()):
    label.set(fontsize=12, color=font_color, **hfont)

plt.subplots_adjust(left=None, bottom=.15, right=None, top=.85, wspace=None, hspace=.9)
```

```python
filename = 'mpl-filled-area'
plt.savefig(filename+'.png', facecolor=facecolor)
```
